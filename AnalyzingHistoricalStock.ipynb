# Use yfinance to Extract Stock Data

```python
# Import the yfinance library
import yfinance as yf

# Create the ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract historical information and save it in a dataframe
tesla_data = tesla.history(period="max")

# Reset the index of the dataframe
tesla_data.reset_index(inplace=True)

# Display the first five rows of the dataframe
tesla_data.head()
```

# Use Webscraping to Extract Tesla Revenue Data

```python
# Get the financials data
tesla_financials = tesla.financials

# Extract the revenue data
tesla_revenue = tesla_financials.loc['Total Revenue']

# Convert the revenue data to a DataFrame
tesla_revenue_df = tesla_revenue.reset_index()
tesla_revenue_df.columns = ['Date', 'Revenue']

# Display the last five rows of the dataframe
tesla_revenue_df.tail()
```

```python
# Create the ticker object for GameStop
gme = yf.Ticker("GME")

# Get the quarterly financials data
gme_quarterly_financials = gme.quarterly_financials

# Extract the revenue data
gme_revenue = gme_quarterly_financials.loc['Total Revenue']

# Convert the revenue data to a DataFrame
gme_revenue_df = gme_revenue.reset_index()
gme_revenue_df.columns = ['Date', 'Revenue']

# Display the last five rows of the dataframe
gme_revenue_df.tail()
```

# Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt

# Function to make the graph
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(title)
    plt.legend()
    plt.show()

# Create the ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract historical stock data and save it in a dataframe
tesla_data = tesla.history(period="max")

# Reset the index of the dataframe
tesla_data.reset_index(inplace=True)

# Plot the Tesla stock data
make_graph(tesla_data, 'Tesla Stock Price Over Time')
```
